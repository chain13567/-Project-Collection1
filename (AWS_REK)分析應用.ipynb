# 數據統計

## 建立 DataFrame

```python
df = pd.DataFrame()
```

## 載入屬性 list 自 DataFrame 中

```python
df['AgeLow'] = AgeLow
df['AgeHigh'] = AgeHigh
df['性別'] = Gender
df['鬍子'] = Beard
df['小鬍子'] = Mustache
df['情緒'] = Emotion
df['戴眼鏡'] = Eyeglasses
df['戴太陽眼鏡'] = Sunglasses
```

```python
df['視覺年齡均值'] = (df['AgeLow']+df['AgeHigh'])/2
```
