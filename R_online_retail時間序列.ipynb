# 清洗資料

## ignore cancel orders 數量為負數，視為取消訂單，即忽略(刪除)

```r
# ignore cancel orders
# 扣除取消訂單
dim(df)
df <- df[which(df$Quantity > 0),  ]
dim(df)
#數量為負數，視為取消訂單，即忽略(刪除)
```

# 時間序列

## 分群:年,月

```r
df_time <- df %>% group_by(year,month) %>% summarise(sum(revenue))
```

```r
df_time
```

## 建立月份時間序列

```r
ts(df_time$`sum(revenue)`, frequency = 12, start = c(2009, 12))
```

## 分成月份

```r
time_month <- ts(df_time$`sum(revenue)`, frequency = 12, start = c(2009, 12))
time_month 
```

# 將月份改為季度

```r
df_time_season <- df
head(df_time_season)
```

```r
df_time_season$month <- ifelse(df_time_season$month >= 1 & df_time_season$month <= 3, 1, df_time_season$month)
df_time_season$month <- ifelse(df_time_season$month >= 4 & df_time_season$month <= 6, 2, df_time_season$month)
df_time_season$month <- ifelse(df_time_season$month >= 7 & df_time_season$month <= 9, 3, df_time_season$month)
df_time_season$month <- ifelse(df_time_season$month >= 10 & df_time_season$month <= 12, 4, df_time_season$month)
```

```r
names(df_time_season)[names(df_time_season) == "month"] <- "season"
```

```r
head(df_time_season)
```

## 分群年,季度

```r
df_time_season <- df_time_season %>% group_by(year,season) %>% summarise(sum(revenue))
```

```r
df_time_season
```

# 時間序列預測 ARIMA模型  Time Series Forecasting

```r
install.packages("zoo")
```

```r
install.packages("forecast")
```

```r
library(zoo)
library(forecast)
```

```r
start(time_month) 
```

```r
end(time_month) 
```

```r
# 擬合一條直線
plot(time_month)
abline(reg=lm(time_month~time(time_month)), col = "red")
```

```r
boxplot(time_month~cycle(time_month), col = 2:20)
```

```r
# 利用auto.arima 
auto.arima(log(time_month),trace= T)
# ARIMA(2,0,2)(0,1,0)[12] with drift         : Inf
# ARIMA(0,0,0)(0,1,0)[12] with drift         : 5.406604
# ARIMA(1,0,0)(0,1,0)[12] with drift         : 8.757287
# ARIMA(0,0,1)(0,1,0)[12] with drift         : 8.77304
# ARIMA(0,0,0)(0,1,0)[12]                    : 2.570779
# ARIMA(1,0,1)(0,1,0)[12] with drift         : 13.08801

# Best model: ARIMA(0,0,0)(0,1,0)[12]                    

# Series: log(time_month) 
# ARIMA(0,0,0)(0,1,0)[12] 

# sigma^2 = 0.05966:  log likelihood = -0.1
# AIC=2.21   AICc=2.57   BIC=2.77
```

```r
auto.arima(time_month,trace= T)
```

```r
# auto.arima提供的最佳模型為ARIMA(0,0,0)(0,1,0)[12] ，我們可以同時測試兩個模型，看看哪個更適合。
airarima1 <- arima(log(time_month),order=c( 0 , 0 , 0 ),
          seasonal=list(order=c( 0 , 1 , 0 ),period= 12 ), 
          method= "ML" )  # 'method' should be one of “CSS-ML”, “ML”, “CSS”
airarima1
```

```r
airarima2 <- arima(time_month,order=c( 0 , 0 , 0 ),
          seasonal=list(order=c( 0 , 1 , 0 ),period= 12 ), 
          method= "ML" )  # 'method' should be one of “CSS-ML”, “ML”, “CSS”
airarima2
```
